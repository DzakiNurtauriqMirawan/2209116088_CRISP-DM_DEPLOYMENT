####Outliers Values

```python
import pandas as pd

# Load data
df = pd.read_csv('CarRentalData.csv')

# Fungsi untuk mendeteksi outliers menggunakan metode IQR (Interquartile Range)
def detect_outliers(data):
    Q1 = data.quantile(0.25)
    Q3 = data.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (data < lower_bound) | (data > upper_bound)
    return outliers

# Deteksi outliers untuk setiap kolom numerik
outliers_df = df.select_dtypes(include=['int64', 'float64']).apply(detect_outliers)

# Hitung persentase outliers
outliers_percentage = (outliers_df.sum() / len(df)) * 100

# Tampilkan hasil
print("Persentase Outliers dalam Dataset:")
print(outliers_percentage)
```

1. **Rating**: Dalam konteks dataset ini, rating mengacu pada peringkat atau penilaian yang diberikan kepada suatu objek atau layanan. #### ROC dan AUC

```python
from sklearn.metrics import roc_auc_score, roc_curve, auc
from sklearn.preprocessing import label_binarize

models = [knn, gnb, dtc]
model_names = ['K-Nearest Neighbor', 'Gaussian Naive Bayes', 'Decision Tree Classifier']

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for model, name, ax in zip(models, model_names, axes):
    y_score = model.predict_proba(x_test_norm)
    y_test_bin = label_binarize(y_test, classes=[0, 1, 2])

    fpr = dict()
    tpr = dict()
    roc_auc = dict()
    for i in range(len(np.unique(y_test))):
        fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    colors = ['darkorange', 'blue', 'green']
    for i in range(len(np.unique(y_test))):
        ax.plot(fpr[i], tpr[i], color=colors[i], lw=2, label=f'Class {i} (ROC-AUC = {roc_auc[i]:.2f})')

    ax.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title(f'ROC Curve - {name}')
    ax.legend(loc='lower right')

plt.tight_layout()
plt.show()
```

Kurva ROC AUC (Receiver Operating Characteristic Area Under the Curve) adalah grafik yang menunjukkan hubungan antara tingkat positif sejati (TPR) dan tingkat positif palsu (FPR) pada berbagai ambang batas klasifikasi.TPR adalah proporsi data positif yang diklasifikasikan dengan benar sebagai positif. FPR adalah proporsi data negatif yang diklasifikasikan dengan benar sebagai negatif AUC (Area Under the Curve) adalah area di bawah kurva ROC. AUC yang lebih tinggi menunjukkan kinerja model klasifikasi yang lebih baik.Bagian X-axis grafik menunjukkan tingkat positif palsu (FPR). Bagian Y-axis grafik menunjukkan tingkat positif sejati (TPR).


Berdasarkan gambar, dapat disimpulkan bahwa:
KNN memiliki AUC tertinggi, diikuti oleh DTC dan GNB.
KNN memiliki kinerja klasifikasi yang lebih baik secara keseluruhan dibandingkan dengan GNB dan DTC.

### Cross Validations

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_validate
from sklearn.metrics import accuracy_score

# Inisialisasi model
models = [gnb, knn, dtc]
model_names = ['Gaussian Naive Bayes', 'K-Nearest Neighbors', 'Decision Tree Classifier']

# Lakukan validasi silang untuk setiap model
cv_results = []
for model, name in zip(models, model_names):
    cv_result = cross_validate(model, x_train_norm, y_train, cv=5, scoring='accuracy', return_train_score=True)
    cv_results.append(cv_result)

# Ekstraksi skor validasi dari hasil
train_scores = []
test_scores = []
for cv_result in cv_results:
    train_scores.append(cv_result['train_score'])
    test_scores.append(cv_result['test_score'])

# Buat dataframe dari hasil validasi silang
df_cv_scores = pd.DataFrame(test_scores, index=model_names).T

# Tampilkan visualisasi hasil validasi silang dengan lineplot
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_cv_scores, markers=True)
plt.title('Cross-Validation Scores for Different Models')
plt.xlabel('Fold')
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend(title='Model', loc='lower right')
plt.xticks(ticks=range(5), labels=['Fold 1', 'Fold 2', 'Fold 3', 'Fold 4', 'Fold 5'])
plt.show()
```

# Tuning Hypermeter

### **GNB**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import GaussianNB

# Definisikan ruang parameter
param_grid = {
    'var_smoothing': np.logspace(0, -9, num=100)
}

# Inisialisasi GridSearchCV dengan model Gaussian Naive Bayes
grid_search = GridSearchCV(estimator=GaussianNB(), param_grid=param_grid, cv=5)

# Melakukan penelusuran grid untuk mencari parameter terbaik
grid_search.fit(x_train_selected, y_train)

# Menampilkan parameter terbaik yang ditemukan
print("Parameter terbaik:", grid_search.best_params_)

# Menggunakan model sebelum tuning untuk menghitung akurasi
gnb_before_tuning = GaussianNB().fit(x_train_selected, y_train)
acc_before_tuning = gnb_before_tuning.score(x_test_selected, y_test)
print("Akurasi sebelum tuning:", acc_before_tuning)

# Menggunakan model terbaik yang ditemukan dari GridSearchCV
best_gnb = grid_search.best_estimator_

# Menggunakan model terbaik untuk menghitung akurasi setelah tuning
acc_after_tuning = best_gnb.score(x_test_selected, y_test)
print("Akurasi setelah tuning:", acc_after_tuning)
```

1. **Parameter terbaik:** Querry tersebut memberikan informasi tentang parameter terbaik yang ditemukan selama proses tuning model. Dalam hal ini, parameter terbaik adalah `{'var_smoothing': 1.0}`. Ini menunjukkan bahwa nilai `var_smoothing` sebesar 1.0 memberikan kinerja model yang optimal setelah tuning.

2. **Akurasi sebelum dan setelah tuning:** Querry tersebut juga memberikan informasi tentang akurasi model sebelum dan setelah proses tuning. Dalam kasus ini, akurasi model sebelum tuning adalah 0.7208, sedangkan akurasi model setelah tuning juga adalah 0.7208. Hal ini menunjukkan bahwa proses tuning tidak mempengaruhi akurasi model. Meskipun parameter telah disesuaikan, kinerja model tidak mengalami perubahan yang signifikan dalam hal akurasi.

### **KNN**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

# Tentukan parameter grid untuk KNN
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9],  # Jumlah tetangga terdekat yang akan dipertimbangkan
    'weights': ['uniform', 'distance'],  # Bobot yang akan digunakan dalam prediksi
    'p': [1, 2]  # Parameter jarak (1 untuk Manhattan, 2 untuk Euclidean)
}

# Inisialisasi model KNN
knn = KNeighborsClassifier()

# Mencari parameter terbaik dengan GridSearchCV
grid_search_knn = GridSearchCV(estimator=knn, param_grid=param_grid_knn, cv=5)
grid_search_knn.fit(x_train_selected, y_train)

# Print parameter terbaik
print("Parameter terbaik untuk KNN:", grid_search_knn.best_params_)

# Menghitung akurasi sebelum tuning
acc_before_tuning_knn = grid_search_knn.best_estimator_.score(x_test_selected, y_test)
print("Akurasi KNN sebelum tuning:", acc_before_tuning_knn)

# Menggunakan model terbaik untuk menghitung akurasi setelah tuning
acc_after_tuning_knn = grid_search_knn.best_estimator_.score(x_test_selected, y_test)
print("Akurasi KNN setelah tuning:", acc_after_tuning_knn)
```

Querry tersebut menyajikan informasi penting terkait tuning model KNN:

1. **Parameter terbaik untuk KNN**: Hasil tuning menunjukkan bahwa parameter terbaik yang diidentifikasi untuk model KNN adalah {'n_neighbors': 9, 'p': 1, 'weights': 'distance'}. Ini berarti model KNN mencapai kinerja optimal dengan menggunakan 9 tetangga terdekat, menggunakan metrik jarak Manhattan (p = 1), dan memberikan bobot jarak berdasarkan jarak antara titik-titiknya.

2. **Akurasi KNN sebelum dan setelah tuning**: Diberikan akurasi sebelum dan sesudah tuning. Pada kasus ini, akurasi KNN sebelum tuning adalah 0.7105, dan setelah tuning tetap sama, yaitu 0.7105. Hal ini menunjukkan bahwa proses tuning tidak mempengaruhi akurasi model KNN. Meskipun parameter telah diubah, kinerja model KNN tidak mengalami perubahan yang signifikan dalam hal akurasi.

### **DTC**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

# Tentukan parameter grid untuk DTC
param_grid_dtc = {
    'criterion': ['gini', 'entropy'],  # Kriteria untuk pemilihan fitur
    'max_depth': [None, 5, 10, 15],  # Kedalaman maksimum dari pohon
    'min_samples_split': [2, 5, 10],  # Jumlah sampel minimum yang diperlukan untuk membagi simpul internal
    'min_samples_leaf': [1, 2, 4]  # Jumlah sampel minimum yang diperlukan di simpul daun
}

# Inisialisasi model DTC
dtc = DecisionTreeClassifier()

# Mencari parameter terbaik dengan GridSearchCV
grid_search_dtc = GridSearchCV(estimator=dtc, param_grid=param_grid_dtc, cv=5)
grid_search_dtc.fit(x_train_selected, y_train)

# Print parameter terbaik
print("Parameter terbaik untuk DTC:", grid_search_dtc.best_params_)

# Menghitung akurasi sebelum tuning
acc_before_tuning_dtc = grid_search_dtc.best_estimator_.score(x_test_selected, y_test)
print("Akurasi DTC sebelum tuning:", acc_before_tuning_dtc)

# Menggunakan model terbaik untuk menghitung akurasi setelah tuning
acc_after_tuning_dtc = grid_search_dtc.best_estimator_.score(x_test_selected, y_test)
print("Akurasi DTC setelah tuning:", acc_after_tuning_dtc)
```

Querry tersebut memberikan penjelasan terkait hasil tuning model Decision Tree Classifier (DTC):

1. **Parameter terbaik untuk DTC**: Informasi ini menjelaskan parameter terbaik yang diidentifikasi selama proses tuning model DTC. Dalam kasus ini, parameter terbaik adalah {'criterion': 'gini', 'max_depth': 5, 'min_samples_leaf': 1, 'min_samples_split': 2}. Ini mengindikasikan bahwa model DTC mencapai kinerja optimal dengan menggunakan kriteria pemisahan 'gini', batasan kedalaman maksimum sebesar 5, jumlah sampel minimum di setiap daun sebesar 1, dan jumlah sampel minimum yang diperlukan untuk membagi simpul sebesar 2.

2. **Akurasi DTC sebelum dan setelah tuning**: Selanjutnya, hasil querry juga menyajikan akurasi model DTC sebelum dan setelah proses tuning. Dalam kasus ini, akurasi DTC sebelum tuning adalah 0.7162, dan setelah tuning tetap sama, yaitu 0.7162. Hal ini menunjukkan bahwa proses tuning tidak mempengaruhi akurasi model DTC. Meskipun parameter telah disesuaikan, kinerja model DTC tidak mengalami perubahan yang signifikan dalam hal akurasi.
